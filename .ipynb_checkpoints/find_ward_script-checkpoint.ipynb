```python
'''
Getting ward for a given address, the wards are the conformed dimension. That means is the column where we can
connect each of the datasets together, each share a Ward. You can look up on what Ward you are in by typing in your
address on a website. We need to refomat some columns so it is suitable to type in. 

Then we can visually explore the data

''' 
from selenium import webdriver
import time
import pandas as pd
import numpy as np

## importing data
col_list = ["_id", "DATE_SUBMITTED", "STREET_DIRECTION", "STREET_NAME","STREET_NUM", "STREET_TYPE"]
df = pd.read_csv('data/csv/Development_Applications_Data.csv', usecols=col_list)  
df["Address"] = df["STREET_NUM"] + " " + df["STREET_NAME"] + " " + df["STREET_TYPE"] + " " + df["STREET_DIRECTION"]

df
```

```python
## Ideally we would have access to an API however we are going to have to hack it
## Warning we have constrainted the number of requests as we 
'''
Ideally we would want to use API to call each 

However we want to do this in an automated fashion

to get the Ward we need to use some growth hacking
tactics and automate
'''

browser = webdriver.Chrome()
browser.get("https://www.toronto.ca/city-government/data-research-maps/neighbourhoods-communities/ward-profiles/")

searcheElement = browser.find_element_by_css_selector("#js_input__address")
array = []
for element in addresses:
    searcheElement.send_keys(element)
    searcheElement.submit() 
    time.sleep(1)
    try:
        linkElement = browser.find_element_by_css_selector(".map--result-text li:nth-child(1) > a")
        linkElement.click()
        searcheElement.clear() 
    except:
        try :
            wardNumElement = browser.find_element_by_css_selector(".map--ward-number")
            array.append(wardNumElement.text)
            searcheElement.clear()
            time.sleep(1)
        except:
            print("An exception occurred - Ward cannot be found")
            array.append("0")
```
