```python
import PyPDF2
import os
import re
import csv

## types of KPIs we are interested in scraping from PDFs on Economic data
## when converted to pdf we use regular expressions to match the metrics
patterns = ["Ward Population\n \d*,\d*","Median Age \d*.\d*", "Ward \d\d|Ward \d"]

def findPattern(patterns, text):
    excel = []
    for pattern in patterns:
        result = re.search(pattern, text)
        if result:
            print("Search successful, found: "+ pattern)
            cell = result.group(0)
            excel.append(cell)
        else:
            print("Search unsuccessful, did not find "+ pattern)
    return excel

# Writing to a csv for easy extraction
def writeCsv(data):
    with open('pdf_census.csv', 'w', newline='') as file:
        writer = csv.writer(file)
        writer.writerow(['Ward Population', 'Median Age', 'Ward'])
        for data in data:
            writer.writerow(data)
            
## Iterating through folder of pdfs, each one representing a ward
def iteratingFolder(directory):
    excel = []
    for filename in os.listdir(directory):
        try: 
            if  filename.endswith(".pdf"):
                pdfFile = open(filename, "rb")
                reader = PyPDF2.PdfFileReader(pdfFile)
                ward_text = reader.getPage(0).extractText()
                data = findPattern(patterns,ward_text)
                excel.append(data)
            else:
                continue
        except:
            "Cannot Find Pdfs"
        writeCsv(excel)

direct = os.chdir(r'C:\\Users\\wyatt\\Desktop\\Accenture\\Toronto_Project\\data\\pdf')
iteratingFolder(direct)
```
